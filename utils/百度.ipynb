```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from urllib.parse import quote
from urllib.parse import unquote
from bs4 import BeautifulSoup

import time
import datetime
import requests

import pandas as pd
import re
import json
```

```python
def get_time_stamp(time_string):
    time_array = time.strptime(time_string,"%Y-%m-%d %H:%M:%S")
    return str(int(time.mktime(time_array)))
```

```python
originalurl = "https://www.baidu.com/s?"
```

```python
chrome_options = Options()
#chrome_options.add_argument('--headless')
web = webdriver.Chrome(options=chrome_options, executable_path="./chromedriver_win32/chromedriver.exe") 
```

```python
def saveinfo(info, path):
    with open(path, "wb+") as f:
        tmp = json.dumps(info, ensure_ascii=False)
        f.write(tmp.encode("utf-8"))
        f.close()
    print("{} 成功保存".format(path))
```

```python
search_word = "华为"
page_num = "0"
begin = datetime.date(2019, 1, 31)
end = datetime.date(2019, 6, 30)
```

```python
for i in range((end-begin).days+1):
    start_time = str(begin + datetime.timedelta(days=i))
    end_time = str(begin + datetime.timedelta(days=i+1))
    search_url = originalurl+"wd={}&pn={}&rtt=1&bsst=1&cl=2&tn=news&rsv_dl=ns_pc".format(search_word, page_num)+\
        "&rsv_dl=tb&gpc=stf%3D{}%2C{}%7Cstftype%3D2".format(get_time_stamp(str(start_time)+" 00:00:00"), get_time_stamp(str(end_time)+" 00:00:00"))


    web.get(search_url)
    data = BeautifulSoup(web.page_source)
    news_url = set()
    for x in data.find_all("a"):
        tmp_str = x.get("href")
        if tmp_str is not None and "baijiahao.baidu.com" in tmp_str :
            news_url.add(tmp_str)
    
    #
    res = []
    for i in news_url:
        title = ''
        content = ''
        author = ''
        newsdate = ''
        newstime = ''
        web.get(i)
        time.sleep(2)
        data = BeautifulSoup(web.page_source)
        try:
            title = data.find(class_="article-title").text
            author = data.find(class_="author-name").text
            newsdate = data.find(class_='date').text
            newstime = data.find(class_='time').text
        except:
            print("{} 无标题。".format(i))
        try:
            try:
                content = data.find(class_="article").text
            except:
                content = data.find(class_="content").text
        except:
            print("{} 无内容。".format(i))
        res.append({
            'title': title,
            'content': content,
            'author': author,
            'newsdate': newsdate,
            'newstime': newstime
        })
    saveinfo(res, "./baidu/"+search_word+start_time+end_time+".json")
```
