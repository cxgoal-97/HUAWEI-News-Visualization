```python
import numpy as np
import pandas as pd
import json
import glob
```

```python
from collections import Counter
```

```python
path = 'utils/cut_word/weibo_cutword/'
```

```python
file_list = glob.glob(path+"*.json")
```

```python
stopwords = ['华为', '、','都','：','一个','很','！','款','4','元','吃','=','\u200b','岁','姑姑','思汗','何建华','@','字','➡️','四','穿']
```

```python
hotword = []

for file in file_list:
    with open(file, 'r', encoding='utf-8') as f:
        d = f.read()
        d = Counter(json.loads(d))
        if(len(d.values())==0):
            hotword.append(('无',0))
            #print(file)
        else:
            for i in d.most_common(n=10):
                if i[0] not in stopwords:
                    #print(i)
                    hotword.append(i)
                    break
        f.close()
```

```python
import datetime
```

```python
hotw = []
hotnum = []
```

```python
for i in hotword:
    hotw.append(i[0])
    hotnum.append(i[1])
```
