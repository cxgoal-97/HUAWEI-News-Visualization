```python
import re
pattern = re.compile('\d{1,2}:\d{1,2}')
```

```python
path = "./data/weibo/"

file_list = glob.glob(path+"*.json")

columns_name = ["filename", "积极总数", "消极总数"] + ["P0"+str(i) for i in range(10)] + ["P"+str(i) for i in range(10,24)] +  \
["N0"+str(i) for i in range(10)] + ["N"+str(i) for i in range(10,24)]

res = pd.DataFrame(columns=columns_name)

res['filename'] = [i[13:35] for i in file_list]
#res = res.fillna(0)
```

```python

for index in range(len(file_list)):
    with open(file_list[index], "r", encoding="utf-8") as f:
        d = f.read()
        f.close()
    dd = json.loads(d)
    
    value_array = np.zeros(len(columns_name)-1)
    for i in dd:
        hours = int(pattern.search(i['time']).group()[:2])
        if i['sentiment'][0]['sentiment'] == 2:
            # 积极
            value_array[0] = value_array[0] + 1
            value_array[2+hours] = value_array[2+hours] + 1
        else:
            value_array[1] = value_array[1] + 1
            value_array[2+24+hours] = value_array[2+24+hours] + 1
    res.loc[index][columns_name[1:]] = value_array

res.to_csv('weibo_time.csv', index=False)
```
